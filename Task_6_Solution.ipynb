```python
#Importing the libraries....
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading the dataset
data=sns.load_dataset("iris")
data.head()
```

```python
sns.pairplot(data,hue="species",height=3);
plt.show()
```

```python
sns.heatmap(data.corr(),annot= True);
```

```python
labelEncoder = LabelEncoder()
labelEncoder.fit(data['species'])
data['species'] = labelEncoder.transform(data['species'])
data.info()
```

```python
data.head()
```

```python
inputs = data.drop('species',axis='columns')
inputs.head()
```

```python
inputs.columns
```

```python
target = data['species']
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(inputs, target, random_state=0, train_size=0.7)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

The accuracy of our model is 97.8% which is pretty high.

```python
import sklearn.datasets as datasets
from sklearn import tree

iris=datasets.load_iris()
# Visualize the graph
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(model, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```
